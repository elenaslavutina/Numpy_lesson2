```python
import pandas as pd
a = { "author_id":[1,2,3],
      "author_name":["Тургенев","Чехов","Островский"]}
authors = pd.DataFrame(a)  
   
c = {"author_id":[1,1,1,2,2,3,3],
      "book_title":['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
     "author_price":[450, 300, 350, 500, 450, 370, 290]}
books = pd.DataFrame(c)   
authors_price = pd.merge(authors, books, on='author_id', how='inner')
authors_price
```

```python
d = authors_price.copy()
top5 = d.nlargest(5,"author_price")
top5
```

```python
groupby = authors_price.groupby("author_name")

df1 = groupby["author_price"].mean()
df2 = groupby["author_price"].max()
df3 = groupby["author_price"].min()
authors_price_agg = pd.merge(df1, df2, on='author_name', how='inner')
authors_price_agg = pd.merge(authors_price_agg, df3, on='author_name', how='inner')

authors_price_agg.columns = ['price_mean', 'price_max', 'price_min']
print(authors_price_agg)
```
